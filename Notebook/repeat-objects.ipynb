```python
import splusdata
import getpass
import pandas as pd
from astropy.table import Table
from pathlib import Path
```

```python
ROOT = Path("../iDR3_n4") 
```

```python
# Connecting with SPLUS database
username = input(prompt="Login: ")
password = getpass.getpass("Password: ")
conn = splusdata.connect(username, password)
```

```python
table = Table.read(ROOT / "Good-LD-Halpha-DR3_noFlag_merge-takeoutbad-Final.ecsv", format="ascii.ecsv")
table
```

```python
#Selecting columns
col = ["Field", "ID", "RA", "DEC"]
table_new = table[col]
table_new
```

```python
Query = f"""SELECT spid.ID, upl.ra, upl.dec 
                FROM TAP_UPLOAD.upload AS upl 
                LEFT OUTER JOIN "idr3_vacs"."splus_id" AS spid ON upl.id = spid.ID WHERE splusIDflag = 0"""
```

```python
df = (table_new.to_pandas())
n = int(len(df) / 5000.) + 1
```

```python
df_ = [] # list
j = 0 # counter
d = {} # empty
for i in range(n):
    j += 1  
    df_.append(df.iloc[5000*i:5000*j])
```

```python
merged_table_list = []
for a in range(2):
    results = conn.query(Query, df_[a])
    merged_table_list.append(results)
```
