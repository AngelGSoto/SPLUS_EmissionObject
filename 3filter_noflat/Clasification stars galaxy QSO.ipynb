```python
from splusdata.features.g_star import ClassifyObj
import pandas as pd
from astropy.table import Table
from pathlib import Path
ROOT_PATH = Path("star-class")
```

```python
tab = pd.read_csv(ROOT_PATH / 'table-version-incom.csv')
tab.columns
```

```python
clf = ClassifyObj(tab, model="RF16", return_prob = True, match_irsa=False)
```

```python
clf.results
```

```python
df = clf.results
```

```python
tab_f = pd.concat([tab, df], axis=1)
```

```python
tab_f
```

```python
asciifile = "table-version-incom-class.ecsv"
Table.from_pandas(tab_f).write(ROOT_PATH / asciifile, format="ascii.ecsv") #, overwrite=True)
```
