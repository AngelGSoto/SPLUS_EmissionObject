```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.table import Table, vstack
from pathlib import Path
import seaborn as sns
import glob
import json
%matplotlib inline
sns.set_color_codes()
```

```python
ROOT_PATH = Path("..") 
```

```python
table1 = Table.read("Halpha-DR3-SPLUS-PStotal-STAR-r16-v2.ecsv", format="ascii.ecsv")
table2 = Table.read("Halpha-DR3-SPLUS-PStotal-STAR-16r18-v2-v2.ecsv", format="ascii.ecsv")
table3 = Table.read("Halpha-DR3-SPLUS-PStotal-STAR-18r20-v2-v2.ecsv", format="ascii.ecsv")
table4 = Table.read("Halpha-DR3-SPLUS-PStotal-STAR-20r21-v2-v2.ecsv", format="ascii.ecsv")
```

```python
# Merge tall four tables
table_merge = vstack([table1, table2, table3, table4])
table_merge.colnames
```

```python
table_merge
```

```python
table_merge.remove_columns(['Unnamed: 0'])
```

```python
table_merge.write("Halpha-DR3_PStotal-STAR_merge.ecsv", format="ascii.ecsv")
table_merge.write("Halpha-DR3_PStotal-STAR_merge.dat", format="ascii.commented_header")
```

```python
#Save as csv file
df = table_merge.to_pandas()
df.to_csv("Halpha-DR3_PStotal-STAR_merge.csv", index=False)
```
